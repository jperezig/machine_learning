```python
import numpy as np
import scipy.special
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
x = np.arange(-3,3,0.01)
y_tanh = np.tanh(x)
d_y_tanh = 1 - np.square(np.tanh(x))
y_sigmoid = scipy.special.expit(x)
plt.plot(x, y_tanh)
plt.plot(x, d_y_tanh)
```

```python
x = np.arange(-3,3,0.01)
y_sigmoid = scipy.special.expit(x)
d_y_sigmoid = scipy.special.expit(x) * (1 - scipy.special.expit(x))
plt.plot(x, y_sigmoid)
plt.plot(x, d_y_sigmoid)
```
