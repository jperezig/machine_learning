```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

%matplotlib inline
```

```python
xx, yy = np.meshgrid(np.linspace(-3,3,50), np.linspace(-3,3,50))
```

```python
rng = np.random.RandomState(0)
X = rng.randn(200, 2)
Y = np.logical_xor(X[:, 0] > 0, X[:, 1] > 0)
```

## 2 Dimensions

```python
plt.figure(figsize=(6,6))
plt.scatter(X[Y][:,0], X[Y][:,1])
plt.scatter(X[np.logical_not(Y)][:,0], X[np.logical_not(Y)][:,1])
plt.axhline(0, color='black')
plt.axvline(0, color='black')
```

## Move to a higher dimensional space

## 3 Dimension

```python
Z = np.logical_and(X[:, 0] > 0, X[:, 1] > 0)
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')
#Z = X[:,0] * X[:,1]
ax.scatter(X[Y][:,0], 
           X[Y][:,1], 
           Z[Y])
ax.scatter(X[np.logical_not(Y)][:,0], 
           X[np.logical_not(Y)][:,1],
           Z[np.logical_not(Y)])
# rotate the axes and update
#for angle in range(0, 360):
#    ax.view_init(70, angle)
#   plt.pause(.001)
#plt.axhline(0, color='black')
#plt.axvline(0, color='black')
#ax.scatter(xs=X_y[X_y[:,3]==0][:,0], ys=X_y[X_y[:,3]==0][:,1], zs=X_y[X_y[:,3]==0][:,2])
#ax.scatter(xs=X_y[X_y[:,3]==1][:,0], ys=X_y[X_y[:,3]==1][:,1], zs=X_y[X_y[:,3]==1][:,2])
```

```python
plt.figure(figsize=(6,6))
plt.scatter(X[Y][:,1], Z[Y])
plt.scatter(X[np.logical_not(Y)][:,1], Z[np.logical_not(Y)])
plt.axhline(0, color='black')
plt.axvline(0, color='black')
```

```python
X = np.c_[xx.ravel(), yy.ravel()]
Z = np.array([0, 1, 1, 0]).reshape(2,2)
```

```python
def score(x):
    return (x[0],x[1],x[1]*x[0])
kkk = np.apply_along_axis(score,1,X)
kkk
```

```python
X_y = np.hstack((kkk, Z.ravel().reshape(4,1)))
```

```python
X_y[X_y[:,3]==0]
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(xs=X_y[X_y[:,3]==0][:,0], ys=X_y[X_y[:,3]==0][:,1], zs=X_y[X_y[:,3]==0][:,2])
ax.scatter(xs=X_y[X_y[:,3]==1][:,0], ys=X_y[X_y[:,3]==1][:,1], zs=X_y[X_y[:,3]==1][:,2])
```
